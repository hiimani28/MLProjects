```python
X = main_data[['_golden', '_unit_state', '_trusted_judgments',
       'gender:confidence', 'profile_yn', 'profile_yn:confidence',
       'fav_number', 'gender_gold', 'link_color', 'profile_yn_gold',
       'retweet_count', 'sidebar_color','tweet_count',
       'tweet_id']].values
Y = main_data[['gender']].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size= 0.2)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, Y_train)
```

```python
Y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```
